```python
from matplotlib import pyplot as plt
from numpy import*
from matplotlib import style
#style.use("ggplot")
x=[1,2,3,4]
y=[11,23,78,34]
x1=[11,22,33,44]
y1=[33,21,45,43]
plt.plot(x,y,"g",label="line one",linewidth=5)
plt.plot(x1,y1,"c",label="line 2",linewidth=5)
plt.title("graph between x-axis and y-axis")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True,color="k")
plt.show()
```

```python
from matplotlib import pyplot as plt
from numpy import*
x=[1,2,12,32]
y=[12,34,32,45]
x1=[21,34,54,76]
y1=[21,87,78,90]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph between x-axis and y-axis")
plt.plot(x,y,"c",label="one line",linewidth=5)
plt.plot(x1,y1,"r",label="second line",linewidth=5)
plt.legend()
plt.grid(True,color="g")
plt.show()
```

```python
x=arange(1,10)
y=(2**x)+10
y1=1230-15*x
plt.title("graph between x-axis and y-axis")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,"c",linewidth=5,label="first graph")
plt.plot(x,y1,"g",linewidth=5,label="secondgraph")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
days=[1,2,3,4,5]
sl=[7,2,2,13]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.pie(sl,labels=[sleeping,eating,working,playing],colors=["c","m","r","b","y"],startangle=90,shadow=True,explode=(0,0.1,0,0),autopct="%1.1f%%")
plt.show()
```

```python
from tkinter import *
#from tkinter.ttk import *

import matplotlib
#matplotlib.use("TkAgg")
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

root = Tk()

figure = Figure(figsize=(5, 4), dpi=150)
plot = figure.add_subplot(1, 1, 1)

plot.plot(0.5, 0.3, color="red", marker="o", linestyle="")

x = [ 0.1, 0.2, 0.3 ]
y = [ -0.1, -0.2, -0.3 ]
plot.plot(x, y, color="blue", marker="x", linestyle="")

canvas = FigureCanvasTkAgg(figure, root)
canvas.get_tk_widget().grid(row=0, column=0)

root.mainloop()
```
